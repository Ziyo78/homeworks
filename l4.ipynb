**Kendall correlation**

En :
Kendalls Tau Coefficient is a non-parametric statistic used to measure the strength and direction of association between two variables based on the ranks of the data. It evaluates how well the orderings of the two variables match by counting the number of concordant and discordant pairs. A value of +1 indicates perfect agreement, -1 indicates perfect disagreement, and 0 suggests no association. Kendall's Tau is especially suitable for ordinal data, or when the relationship between variables is non-linear or the data is not from a normal distribution.

Ru :       ,              ranks.      ,   concordant  discordant .  +1   , -1   ,  0   .        ordinal data,       non-linear     normal distribution.

Uz : Kendall Tau koeffitsienti  bu ikki ozgaruvchining ozaro bogliqligini ranks asosida baholovchi parametrik bolmagan statistik korsatkichdir. Bu korsatkich ozgaruvchilarning tartibda qanchalik mos kelishini concordant va discordant juftliklar soni orqali baholaydi. +1 qiymati toliq moslikni, -1 toliq zidlikni, 0 esa bogliqlik yoqligini anglatadi. Kendall Tau ayniqsa ordinal data, non-linear munosabatlar yoki normal distribution ga ega bolmagan malumotlar bilan ishlashda foydalidir.

```python
def kendal(df, colum1, colum2):
    index = 0
    df = df.sort_values(colum1)
    df.reset_index(inplace=True)
    count_plus = 0
    count_minus = 0
    for i in df[colum2]:
        for j in df[colum2].loc[index+1:]:
            if j > i:
                count_plus += 1
            else:
                count_minus += 1
        index += 1
    return (count_plus - count_minus) / (count_plus + count_minus)
print(kendal(ex, 'Time of study', 'Exam Score'))
```
