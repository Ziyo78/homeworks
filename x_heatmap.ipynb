```python
from ucimlrepo import fetch_ucirepo
import matplotlib.pyplot as plt
import numpy as np
```

```python
heart_disease = fetch_ucirepo(id=45)

X = heart_disease.data.features
Y = heart_disease.data.targets
print(heart_disease.metadata)
print(heart_disease.variables)
```

```python
plt.matshow(X.corr(), cmap = 'coolwarm')

plt.xticks(ticks=np.arange(len(X.corr().columns)), labels=X.corr().columns, rotation=90)
plt.yticks(ticks=np.arange(len(X.corr().columns)), labels=X.corr().columns)

plt.colorbar()

plt.tight_layout()
plt.show()
```
